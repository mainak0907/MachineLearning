Finding Outlayers 

```python
df['V203523'].hist(grid=False,figsize=(20, 10))
```

```python
outliers=[]
def detect_outlier(data_1):
    
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers


outlier_datapoints = detect_outlier(df['V200002'])
print(outlier_datapoints)


df5 = pd.DataFrame(outlier_datapoints)
df5.hist()
```
